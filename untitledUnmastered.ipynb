```python
import torch
from torchvision import datasets, transforms

transform = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.ToTensor()
])

trainval_dataset = datasets.OxfordIIITPet(
    root='./data',
    split='trainval',
    download=True,
    transform=transform
)

test_dataset = datasets.OxfordIIITPet(
    root='./data',
    split='test',
    download=True,
    transform=transform
)

print("Trainval size:", len(trainval_dataset))
print("Test size:", len(test_dataset))
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.transforms.functional as TF
import random
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, random_split, Subset
```

```python
class PetWithTrimapCrop(torch.utils.data.Dataset):
    def __init__(self, root, split, download=False, augment=False):
        self.dataset = datasets.OxfordIIITPet(
            root=root,
            split=split,
            target_types=['category', 'segmentation'],
            download=download,
        )
        self.augment = augment
        self.normalize = transforms.Normalize(
            mean=[0.47828054428100586, 0.4458567202091217, 0.3956858813762665],
            std=[0.22214548289775848, 0.2190534770488739, 0.22055834531784058]
        )

    def __len__(self):
        return len(self.dataset)

    def _get_bbox_from_trimap(self, segmentation):
      mask = TF.to_tensor(segmentation)
      foreground = (mask == 1/255)
      if foreground.sum() == 0:
        return None
      cols = foreground.any(dim=1).squeeze()
      rows = foreground.any(dim=2).squeeze()
      ymin = rows.nonzero()[0].item()
      ymax = rows.nonzero()[-1].item()
      xmin = cols.nonzero()[0].item()
      xmax = cols.nonzero()[-1].item()

      # Add padding around the crop
      h, w = mask.shape[1], mask.shape[2]
      box_w = xmax - xmin
      box_h = ymax - ymin
      pad_x = int(box_w * 0.11)  # 10% padding
      pad_y = int(box_h * 0.11)
      xmin = max(0, xmin - pad_x)
      ymin = max(0, ymin - pad_y)
      xmax = min(w, xmax + pad_x)
      ymax = min(h, ymax + pad_y)
      return xmin, ymin, xmax, ymax

    def __getitem__(self, idx):
        image, (category, segmentation) = self.dataset[idx]

        if self.augment:
            angle = random.uniform(-12, 12)
            image = TF.rotate(image, angle)
            segmentation = TF.rotate(segmentation, angle,
                                  interpolation=TF.InterpolationMode.NEAREST)

        bbox = self._get_bbox_from_trimap(segmentation)
        if bbox is not None:
            xmin, ymin, xmax, ymax = bbox
            image = TF.crop(image, ymin, xmin, ymax - ymin, xmax - xmin)

        image = TF.resize(image, [128, 128])

        if self.augment:
            if random.random() > 0.5:
                image = TF.hflip(image)
            image = transforms.ColorJitter(
                brightness=0.25, contrast=0.25, saturation=0.25)(image)

        img_tensor = TF.to_tensor(image)
        img_tensor = self.normalize(img_tensor)
        return img_tensor, category

# Create datasets
# When adding val split back change train_dataset too train_val_aug
train_dataset = PetWithTrimapCrop('./data', split='trainval', augment=True)
#train_val_noaug = PetWithTrimapCrop('./data', split='trainval', augment=False)
test_dataset  = PetWithTrimapCrop('./data', split='test', augment=False)

# Val split
#torch.manual_seed(42)
#n = len(train_val_aug)
#indices = torch.randperm(n).tolist()
#train_indices = indices[:int(0.8 * n)]
#val_indices   = indices[int(0.8 * n):]

#train_dataset = Subset(train_val_aug,   train_indices)
#val_dataset   = Subset(train_val_noaug, val_indices)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
#val_loader   = DataLoader(val_dataset,   batch_size=64, shuffle=False)
test_loader  = DataLoader(test_dataset,  batch_size=64, shuffle=False)

print("Train:", len(train_dataset))
#print("Val:",   len(val_dataset))
print("Test:",  len(test_dataset))
```

```python
class PetCNN(nn.Module):
    def __init__(self):
        super(PetCNN, self).__init__()

        self.features = nn.Sequential(
            # Block 1
            nn.Conv2d(3, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),  # 64x64 -> 32x32
            #nn.Dropout2d(0.1),

            # Block 2
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),  # 32x32 -> 16x16
            #nn.Dropout2d(0.2),

            # Block 3
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),   # 16x16 -> 8x8
           # nn.Dropout2d(0.3),

            # Block 4
            nn.Conv2d(128, 256, kernel_size=3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),

        )

        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(256 * 8 * 8, 512),
            nn.ReLU(),
            nn.Linear(512, 37)   # 37 breeds
        )

    def forward(self, x):
        x = self.features(x)
        x = self.classifier(x)
        return x

model = PetCNN()
print(model)
print("\nTotal parameters:", sum(p.numel() for p in model.parameters()))
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

num_epochs = 30

scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=num_epochs)

best_val_acc = 0.0

for epoch in range(num_epochs):
    # --- Training ---
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        _, predicted = torch.max(outputs, 1)
        correct += (predicted == labels).sum().item()
        total += labels.size(0)

    train_acc = 100 * correct / total
    avg_loss = running_loss / len(train_loader)

    # --- Validation ---
    #model.eval()
    #val_correct = 0
    #val_total = 0

    #with torch.no_grad():
        #for images, labels in val_loader:
            #images, labels = images.to(device), labels.to(device)
           # outputs = model(images)
            #_, predicted = torch.max(outputs, 1)
           # val_correct += (predicted == labels).sum().item()
            #val_total += labels.size(0)

    #val_acc = 100 * val_correct / val_total

    print(f"Epoch [{epoch+1}/{num_epochs}] "
          f"Loss: {avg_loss:.4f} | "
          f"Train Acc: {train_acc:.2f}%")

     # after the print statement, inside the loop:
    scheduler.step()

torch.save(model.state_dict(), 'pet_cnn_best.pth')
print("Model saved.")
```

```python
model.eval()
test_correct = 0
test_total = 0
with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs, 1)
        test_correct += (predicted == labels).sum().item()
        test_total += labels.size(0)
print(f"Test Accuracy: {100 * test_correct / test_total:.2f}%")
```
